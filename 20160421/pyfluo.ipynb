```python
import pyfluo as pf
import numpy as np
%matplotlib inline
from matplotlib import pyplot as pl
pl.style.use('normal')
```

```python
mov = pf.Movie('purkinje_cells.tif')
```

```python
mov.play(loop=True, rolling_mean=5., contrast=3.5)
```

```python
mov.project(show=True)
```

```python
roi = pf.ROI(np.load('roi.npy'))
mov.project(show=True, roi=roi)
```

```python
traces = mov.extract(roi)
traces
```

```python
dff = pf.compute_dff(traces)
```

```python
dff.plot(stacked=True)
```
